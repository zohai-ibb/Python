### `Problem 1`: Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
salary = float(input("Enter CTC Salary (in lakhs): "))

# Deductions
hra = 0.10 * salary
da = 0.05 * salary
pf = 0.03 * salary

salary_after_deduction = salary - (hra + da + pf)

# Tax calculation based on slab
if salary < 5:
    final_salary = salary_after_deduction

elif salary >= 5 and salary < 10:
    final_salary = salary_after_deduction - (0.10 * salary_after_deduction)

elif salary >= 10 and salary < 20:
    final_salary = salary_after_deduction - (0.20 * salary_after_deduction)

else:
    final_salary = salary_after_deduction - (0.30 * salary_after_deduction)

print("In-hand Salary after all deductions:", final_salary, "lakhs")
````

### `Problem 6` - Find the factorial of a given number.

Write a program to use the loop to find the factorial of a given number.

The factorial (symbol: `!`) means to multiply all whole numbers from the chosen number down to 1.

For example: calculate the factorial of 5

```bash
5! = 5 × 4 × 3 × 2 × 1 = 120
```

Output:

```bash
120
```

````python
# Write code here
n = int(input("Enter the number to calc factorial :"))
fact = 1;

for i in range(n, 0, -1):
  fact = fact * n;
  n = n - 1;
print(fact)


````

###`Problem 11`: A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps.
The trace of robot movement is shown as the following:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
> The numbers after the direction are steps.

> `!` means robot stop there.

**Please write a program to compute the distance from current position after a sequence of movement and original point.**

*If the distance is a float, then just print the nearest integer.*

Example:

`Input`:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
`Output`:
```
2
```

````python
# Write code here
import math

x, y = 0, 0

while True:
    move = input().strip()

    if move == "!":
        break

    direction, steps = move.split()
    steps = int(steps)

    if direction == "UP":
        y += steps
    elif direction == "DOWN":
        y -= steps
    elif direction == "LEFT":
        x -= steps
    elif direction == "RIGHT":
        x += steps

# Distance from origin
distance = math.sqrt(x*x + y*y)

# Print nearest integer
print(round(distance))
````

###`Problem 13`:Print all the Armstrong numbers in a given range.
Range will be provided by the user<br>
Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

````python
for i in range(0, 300):
    x = i
    # Count number of digits
    digits = 0
    temp = x
    if temp == 0:   # special case for 0
        digits = 1
    else:
        while temp > 0:
            temp //= 10
            digits += 1

    total = 0
    temp = x
    while temp > 0:
        digit = temp % 10
        total += digit ** digits
        temp //= 10

    if total == i:
        print(i)
````
