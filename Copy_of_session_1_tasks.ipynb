````python
# Write your code here
import math

x1 = float(input("Enter X1 :"))
x2 = float(input("Enter X2 :"))
y1 = float(input("Enter Y1 :"))
y2 = float(input("Enter Y2 :"))

euclidean_distance = math.sqrt((x2 - x1)**2 + (y2 - y1)**2)
print(euclidean_distance)

````
