```python
f = open("Sample.txt", "r")
s = f.read()
print(s)
f.close()

```

```python
f = open("Sample.txt", "r")
for line in f:
  print(line, end="")

f.close()

```

Using **with**


```python
with open("Sample.txt", "w") as f:
  f.write("Hello World\n")
  f.write("Hello World\n")
  f.write("Hello World\n")
  f.write("Hello World\n")
  f.write("Hello World\n")
  f.write("Hello World\n")
```

```python
with open("Sample.txt", "r") as f:
  print(f.read(10))
  print(f.read(10))
  print(f.read(10))
  print(f.tell())
  f.seek(10)
  print("\n")
  print(f.read(10))
```

```python
#Using Json custom format

import json

class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age

person = Person("Zach", 19)

def show(person):
  if(isinstance(person, Person)):
    return "Name -> {} , Age -> {}".format(person.name, person.age)

with open("demo.json", 'w') as f:
  json.dump(person, f, default=show)


```

```python
import json
with open("demo.json", 'r') as f:
  print(json.load(f))
```

```python
#Using Pickle

class Person:
  def __init__(self, name, age):
    self.name = name
    self.age = age
  def display(self):
    print("Name -> {} Age -> {}".format(self.name, self.age))

person = Person("Zach", 19)

import pickle
with open("person.pkl", 'wb') as f:
  pickle.dump(person, f)

```

```python
with open("person.pkl", 'rb') as f:
  p = pickle.load(f)

p.display()
```
